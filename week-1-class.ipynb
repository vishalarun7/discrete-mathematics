Recursive
Table
Keep last two values
Analytic (if you have time)
Matrix (if you have time)


```python
import math
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python

def F(n):
    a = b = 1
    if n == 1 or n == 2:
        return 1
    for i in range(3, n + 1):
        a, b = b, a + b
    return b
```

```python
def G(n):
    if n == 0: return 0
    phi = (1 + math.sqrt(5)) / 2
    return round((phi ** n - (1 - phi) ** n) / math.sqrt(5))
```

```python

for n in range(2, 100):
    if F(n) != G(n):
        print(f"Mismatch at n = {n}: F(n) = {F(n)}, G(n) = {G(n)}")
        break
```

```python
def timeit(n):
    start = time.time()
    G(n)
    end = time.time()
    return end - start


N = list(range(1, 200))
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Time to compute nth Fibonacci number using Analytic Method')
plt.show()
```

```python
#recursive

def recursive (n):
    if n<=2:
        return 1 
    else:
        return recursive (n-1) + recursive (n-2)

def timeit(n):
    start = time.time()
    recursive(n)
    end = time.time()
    return end - start


N = list(range(1, 10))
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.show()
```

```python
def table(n):
    if n <= 0:
        return []
    if n == 1:
        return [0]
    table = [0, 1]
    for i in range(2, n):
        table.append(table[i-1] + table[i-2])
    return table


def timeit(n):
    start = time.time()
    table(n)
    end = time.time()
    return end - start


N = list(range(1, 200))
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.show()
```

```python
#keep 2

def keep2(n):
    a=b=1
    for i in range(3,n):
        a,b = b, a+b
    return b

def timeit(n):
    start = time.time()
    keep2(n)
    end = time.time()
    return end - start


N = list(range(1, 200))
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.show()
```
